# Выявление определяющих успешность игры закономерностей.


Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нужно<b> выявить определяющие успешность игры закономерности</b>. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Представлены данные до 2016 года. Представим, что сейчас декабрь 2016 г., и я планирую кампанию на 2017-й. Нужно отработать принцип работы с данными.
<font color="grey">
    
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».</font>


## Описание данных

Name — название игры

Platform — платформа

Year_of_Release — год выпуска

Genre — жанр игры

NA_sales — продажи в Северной Америке (миллионы проданных копий)

EU_sales — продажи в Европе (миллионы проданных копий)

JP_sales — продажи в Японии (миллионы проданных копий)

Other_sales — продажи в других странах (миллионы проданных копий)

Critic_Score — оценка критиков (максимум 100)

User_Score — оценка пользователей (максимум 10)

Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Обработка пропусков 

```python
# Привести все строки датафрейма к нижнему регистру

columns = df.columns.tolist()
for i in range(df.shape[1]):
    if df.loc[:,columns[i]].dtype == 'object':
        df.loc[:,columns[i]] = df.loc[:,columns[i]].str.lower()


# Привести все названия столбцов к нижнему регистру

df.columns = [x.lower() for x in df.columns]
#df.columns = df.columns.str.lower()
#df.columns = map(str.lower, df.columns)

#display(df.head())
```

```python
# Обработать пропуски в столбце rating

#display(df['rating'].unique())
df['rating'] = df['rating'].fillna('rating_pending')
df['rating'] = df['rating'].replace(
    {'e' : 'everyone',
    'm' : 'mature_17+',
    't' : 'teen_13+',
    'e10+' : 'everyone_10+',
    'k-a' : 'everyone',
    'ao' : 'adults_only_18+',
    'ec' : 'early_childhood',
    'rp' : 'rating_pending'}
)
#display(df['rating'].unique())
#display(df.head(15))
```

```python
# Пропуски в столбце genre

df['genre'] = df['genre'].fillna('misc')
#display(df['genre'].unique())

# Пропуски в столбце name

df['name'] = df['name'].fillna('no_name')
#display(df.query('name == "no_name"'))

# Пропуски в столбце year_of_release

df['year_of_release'] = df['year_of_release'].fillna(0)

# Пропуски в столбце user_score

df['user_score'] = df['user_score'].fillna(-1)
df['user_score'] = df['user_score'].replace('tbd', -1)
```

### Вывод

<div style="border:groove grey 5px; padding: 20px">
    
Привела все строковые значения к нижнему регистру.
    
Заменила пропуски в столбце rating на rating_pending, что означает "рейтинг не определен".
    
Пропуски в столбце genre заменила на "misc", что означает "разное".
    
Два пропущенных названия заменила на no_name, это какие-то супер старые малоизвестные игры из 93 года, которые на общую картину никак не повлияют.
    
Вместо пропусков в столбцах user_score и year_of_release поставила значения-заглушки, чтобы потом заменить тип данных.
    
Решила не заполнять пропуски в столбце critic_score, потому что это важный показатель, его нельзя заменить медианой или средним.

```python
#  Платформы с наибольшими суммарными продажами

pop_platform = (
    data.pivot_table(index='platform', values='total_sales', aggfunc='sum')
    .sort_values(by='total_sales', ascending=False)
    .head(10)
)

pop_platform.reset_index(inplace=True)
display(pop_platform)
list_pop_platform = pop_platform['platform'].tolist()

# Сколько игр выпускалось в разные годы под платформы с самыми большими уровнями продаж

plt.figure(figsize=(12, 8))
for platform in list_pop_platform:
    df_years = data.loc[df.loc[:,'platform'] == platform]['year_of_release'] 
    
    plt.hist(df_years, bins=(50), alpha = 0.7, label=platform, width = 0.5)
    plt.legend(loc='upper left')
plt.title('Годы выпуска')
plt.show()

# Какой в среднем срок жизни платформы

life_years = []
for platform in list_pop_platform:
    years_platform = data.query('platform== @platform').pivot_table(index=['platform', 'year_of_release'], values='name', aggfunc='count')
    life_years.append(len(years_platform))
display('Срок жизни платформы в среднем: {} лет'.format(sum(life_years)/len(life_years)))
```

### Актуальный период

```python
#  Добавляю платформы, которые не вошли в основной топ продаж, но являются более актуальными на 2016 год:

#  Составляю новый топ на момент 2016 года
pop_platform_in_2016 = (
    data.query('year_of_release > 2014')
    .pivot_table(index='platform', values='total_sales', aggfunc='sum')
    .sort_values(by='total_sales', ascending=False).head(5)
)
pop_platform_in_2016.reset_index(inplace=True)
pop_platform_in_2016_list = pop_platform_in_2016['platform'].tolist()
display(pop_platform_in_2016_list)
display(pop_platform_in_2016)

#  Объединяю списки с названиями платформ из основного топа и топа на 2016 год
for element in pop_platform_in_2016_list:
    if element not in list_pop_platform:
        list_pop_platform.append(element)
        
#display(list_pop_platform)
```

```python
#  Актуальные платформы

start_year = data['year_of_release'].max() - 10

actual_platforms_list = []

for platform in list_pop_platform:
    list_years = data.query('platform == @platform')['year_of_release'].tolist()
    if 2016 in list_years:                           # проверяю наличие выпущенных в 2016 году игр
        actual_platforms_list.append(platform)
        plt.hist(list_years, bins=50, range=(start_year, start_year + 10), alpha = 0.5, label=platform, width = 0.5)
        plt.legend(loc='upper left')
        plt.show()


print(actual_platforms_list)
```

## Портрет пользователя каждого региона

### Популярные платформы и жанры по регионам

```python
#  Платформы с наибольшими суммарными продажами по регионам

region_sales_columns = ['na_sales', 'eu_sales', 'other_sales', 'jp_sales']

pop_region_platform_merge = pd.DataFrame(data=[], columns=['platform'])   # Cоздаем общую таблицу platform

pop_region_genre_merge = pd.DataFrame(data=[], columns=['genre'])      # Cоздаем общую таблицу genre

fig, ax = plt.subplots()

for parametr in ['platform','genre']:
    i=1
    for region_sales in region_sales_columns:                              # Таблица для каждого региона
        pop_region = (
            data.pivot_table(index=parametr, values=region_sales, aggfunc='sum')
            .sort_values(by=region_sales, ascending=False)
            .head(10)
        )
        pop_region.reset_index(inplace=True)
        
        # Считаю доли продаж и записываю их в отдельный столбец
        name_sales_part = region_sales.split('_')
        name_sales_part.append('part')
        pop_region['_'.join(name_sales_part)] = pop_region[region_sales]/data.query('year_of_release > @start_year')[region_sales].sum()
        #display(pop_region)
        
        if parametr == 'platform':                                # Добавляю в общую таблицу
            
            pop_region_platform_merge = pop_region_platform_merge.merge(pop_region, on='platform', how='outer')
            
        if parametr == 'genre':                                   # Добавляю в общую таблицу
           
            pop_region_genre_merge = pop_region_genre_merge.merge(pop_region, on='genre', how='outer')
            if i ==1:
                i=2
                
        # Графики по разным регионам
        ax = plt.subplot(4, 2, i)
        plt.tight_layout()  # Нормесное расстояние между графиками
        
        pop_region.plot(kind='bar', x=parametr, y=region_sales, label=region_sales, ax=ax, alpha=0.5, figsize=(17, 20))
        i+=2
        for row in pop_region.itertuples():  # Добавляю доли на графики
            #print(row)
            plt.text(getattr(row, 'Index')-0.2, getattr(row, region_sales)-1, '{:.1%}'.format(getattr(row, '_'.join(name_sales_part))), alpha=0.9, fontsize=16)
        plt.title(region_sales)
plt.show()   

# Вывожу общие таблицы
#display(pop_region_platform_merge)
#display(pop_region_genre_merge)
```

#### Вывод

<div style="border:groove grey 5px; padding: 20px">

Самые популярные жанры в разных регионах, как правило, совпадают, это жанры: action, sports, shooter (входят в топ 5 в каждом регионе). ## Проверка гипотез

```python
# H0 - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

sample_1 = data.query('platform == "xone"')['user_score']
sample_2 = data.query('platform == "pc"')['user_score']
    
alpha = 0.05
results = st.ttest_ind(sample_1, sample_2)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Средние пользовательские рейтинги платформ Xbox One и PC различаются')
else:
    print('Средние пользовательские рейтинги платформ Xbox One и PC не различаются')
```

```python
# Получаем массив со случайными значениями из выборки с возможностью повторения
def get_sample(data):
    return np.random.choice(data, size=len(data))

def get_distplot(sample_1, sample_2):
    # Берем распределение средних этих массивов
    mean_dist_1 = [np.mean(get_sample(sample_1)) for _ in range (10000)]
    mean_dist_2 = [np.mean(get_sample(sample_2)) for _ in range (10000)]

    # Разница средних
    sns.distplot(np.array(mean_dist_1)) 
    sns.distplot(np.array(mean_dist_2))
    
get_distplot(sample_1, sample_2)
plt.legend(['xone','pc'],loc='upper left')
```

### Вывод

<div style="border:groove grey 5px; padding: 20px">
    
Хоть средние пользовательские рейтинги не одинаковы для Xbox One и PC, но подобное различие можно получить случайно.
    
Слишком большая вероятность, чтобы делать вывод о значимом различии между платформами. 


```python
# H0 - Средние пользовательские рейтинги жанров Action и Sports одинаковые
#data = data.query('year_of_release > @start_year')
sample_a = data.query('genre == "action"')['user_score']
sample_s = data.query('genre == "sports"')['user_score']
    
alpha = 0.05
results = st.ttest_ind(sample_a, sample_s)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Средние пользовательские рейтинги жанров Action и Sports отличаются')
else:
    print('Средние пользовательские рейтинги жанров Action и Sports не отличаются')
```
